```python
import pandas as pd
x_data = pd.read_csv("./x_data.csv")
y_data = pd.read_csv("./y_data.csv")
x_data.to_csv("./x_data.csv",index=False)
y_data.to_csv("./y_data.csv",index=False)
x_data = pd.read_csv("./x_data.csv")
y_data = pd.read_csv("./y_data.csv")
x_data.sample()
```

```python
x_data["HDD"].fillna(0,inplace=True)
x_data["SSD"].fillna(0,inplace=True)
```

```python
x_data["SSD"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data,y_data,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
# algo
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, ExtraTreesRegressor
from sklearn.svm import SVR

```

```python
step1 = ColumnTransformer(
  transformers=[
    ("categroical-data",OneHotEncoder(drop="first"),[0,1,7,10,11])
  ]
,remainder="passthrough")

step2 = LinearRegression()
```

```python
x_data
```

```python
pipe = Pipeline([
  ("step1",step1),
  ("step2",step2)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
print("r2score = ",r2_score(y_test,y_pred))
print("mae = ",mean_absolute_error(y_test,y_pred))
```

```python
step3 = ColumnTransformer(
  transformers=[
    ("categroical-data",OneHotEncoder(drop="first",handle_unknown="ignore"),[0,1,7,10,11])
  ]
,remainder="passthrough")
step4 = RandomForestRegressor(n_estimators=100,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)
pipe2 = Pipeline([
  ("step3",step3),
  ("step4",step4)
])
pipe2.fit(x_train,y_train)
```

```python
y_pred = pipe2.predict(x_test)
from sklearn.metrics import mean_absolute_error,r2_score
print("r2score = ",r2_score(y_test,y_pred))
print("mae = ",mean_absolute_error(y_test,y_pred))
```

```python
import pickle as pk
pk.dump(pipe2,open("./model/model.pkl","wb"))
```
